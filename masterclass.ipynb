```python
import warnings
warnings.filterwarnings('ignore')
from tabulate import tabulate
from transformers import AutoTokenizer, AutoModel, pipeline, set_seed
from ipywidgets import interact, widgets
from IPython.display import display
from IPython.utils import io
```

```python
gpt2_tokenizer = AutoTokenizer.from_pretrained("gpt2")
generator = pipeline('text-generation', model='gpt2', pad_token_id=gpt2_tokenizer.eos_token_id)
```

```python
text = widgets.Text(
    value='This is a great',
    placeholder='Put a sentence in here!',
    description='Start:',
    disabled=False
)

temp_slider = widgets.FloatSlider(
    value=.5,
    min=.1,
    max=3.0,
    step=0.1,
    description='Temp.:',
    readout_format='.1f',
)

length_slider = widgets.IntSlider(
    value=10,
    min=5,
    max=50,
    step=1,
    description='Length:',
)

sample_slider = widgets.IntSlider(
    value=3,
    min=1,
    max=5,
    step=1,
    description='Samples:',
)

@interact
def show_output(text=text, temperature=temp_slider, length=length_slider, samples=sample_slider):
  with io.capture_output() as captured:
    generated_examples = generator(text, truncation=True, max_length=length, num_return_sequences=samples, temperature=temperature)
  output = widgets.HTML(
    value=tabulate(generated_examples, tablefmt='html'),
    placeholder='output...',
    description='ouput',
  )

  display(output)
```
